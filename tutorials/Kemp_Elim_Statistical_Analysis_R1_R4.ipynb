```python
# all input files 
R1_in_dir = r"C:\Users\Rory Crean\Dropbox (lkgroup)\Backup_HardDrive\Postdoc\MachLearnConformationalFeatures\Workup\KE07_Workup\Raw_Datasets\R1_5d2w"
R1_out_dir = r"C:\Users\Rory Crean\Dropbox (lkgroup)\Backup_HardDrive\Postdoc\MachLearnConformationalFeatures\Workup\KE07_Workup\Raw_Results_Stats\R1_5d2w"

R1_pycontact_files_horizontal = ["PyContact_Per_Frame_Interactions_Block1.csv", "PyContact_Per_Frame_Interactions_Block2.csv",
                              "PyContact_Per_Frame_Interactions_Block3.csv", "PyContact_Per_Frame_Interactions_Block4.csv",
                              "PyContact_Per_Frame_Interactions_Block5.csv", "PyContact_Per_Frame_Interactions_Block6.csv",
                              "PyContact_Per_Frame_Interactions_Block7.csv", "PyContact_Per_Frame_Interactions_Block8.csv",
                              "PyContact_Per_Frame_Interactions_Block9.csv", "PyContact_Per_Frame_Interactions_Block10.csv",
                              "PyContact_Per_Frame_Interactions_Block11.csv", "PyContact_Per_Frame_Interactions_Block12.csv",
                              "PyContact_Per_Frame_Interactions_Block13.csv", "PyContact_Per_Frame_Interactions_Block14.csv",
                              "PyContact_Per_Frame_Interactions_Block15.csv", "PyContact_Per_Frame_Interactions_Block16.csv",
                              "PyContact_Per_Frame_Interactions_Block17.csv"]

R1_classifications_file = r"C:\Users\Rory Crean\Dropbox (lkgroup)\Backup_HardDrive\Postdoc\MachLearnConformationalFeatures\Workup\KE07_Workup\Raw_Datasets\R1_5d2w\R1_5d2w_1in10_Classified.dat"


# dataset prep parts. 
R1_pycontact_dataset = pycontact_processing.PyContactInitializer(
    pycontact_files=R1_pycontact_files_horizontal,
    multiple_files=True,
    merge_files_method="horizontal",  
    remove_false_interactions=True,
    in_dir=R1_in_dir,
)

# SPECIAL STEP FOR KEMP ELIMINASES BECAUSE PYCONTACT RENUMBERED THEM TO 0. 
R1_prepared_renumberd_df = pycontact_processing.modify_column_residue_numbers(
    dataset=R1_pycontact_dataset.prepared_df, 
    constant_to_add=1
)

R1_supervised_dataset = data_preperation.SupervisedFeatureData(
    input_df=R1_prepared_renumberd_df,
    target_file=R1_classifications_file,
    is_classification=True,
    header_present=True 
)

R1_supervised_dataset.df_processed
```

```python
R1_supervised_dataset.df_processed.Target.value_counts()
```

```python
# Filtering 
R1_supervised_dataset.reset_filtering()
R1_supervised_dataset.filter_by_occupancy_by_class(min_occupancy=25)
R1_supervised_dataset.filter_by_avg_strength(average_strength_cut_off=1.0)
print(f"Number of features after filtering by average interaction scores: {len(R1_supervised_dataset.df_filtered.columns)}")
```

```python
R1_stat_model = stat_modelling.ClassificationStatModel(
    dataset=R1_supervised_dataset.df_filtered, 
    class_names=["ConfA", "ConfC"], 
    out_dir=R1_out_dir,
    interaction_types_included=["Hbond", "Saltbr", "Hydrophobic", "Other"] 
)
```

```python
R1_stat_model.calc_js_distances()
```

```python
R1_stat_model.calc_mutual_info_to_target()
```

```python
# First generate an instance of the class. 
R1_post_proc = post_proccessing.StatClassificationPostProcessor(
    stat_model=R1_stat_model,
    out_dir=R1_out_dir
)

# Now we can run the get_per_res_importance() method, changing the stat_method accordingly.
R1_js_per_res_importances = R1_post_proc.get_per_res_importance(
    stat_method="jensen_shannon")
R1_mi_per_res_importances = R1_post_proc.get_per_res_importance(
    stat_method="mutual_information")

R1_post_proc.estimate_feature_directions()
```

```python
# Write PyMOL compatable scripts for the per feature results.
# Simply swap between the two statistical methods as shown below. 
pymol_projections.project_pymol_top_features(
    per_feature_scores=R1_stat_model.js_distances,
    model_name="jensen_shannon",
    numb_features=100,
    out_dir=R1_out_dir
)

pymol_projections.project_pymol_top_features(
    per_feature_scores=R1_stat_model.mutual_infos,
    model_name="mutual_information",
    numb_features=100,
    out_dir=R1_out_dir
)

# Write PyMOL compatable scripts for the per residue results.
# Simply swap between the two statistical methods as shown below. 
pymol_projections.project_pymol_per_res_scores(
    per_res_scores=R1_js_per_res_importances,
    model_name="jensen_shannon",
    out_dir=R1_out_dir
)
pymol_projections.project_pymol_per_res_scores(
    per_res_scores=R1_mi_per_res_importances,
    model_name="mutual_information",
    out_dir=R1_out_dir
)
```

#### R1_5d2w

```python
# all input files 
R1_in_dir = r"C:\Users\Rory Crean\Dropbox (lkgroup)\Backup_HardDrive\Postdoc\MachLearnConformationalFeatures\Workup\KE07_Workup\Raw_Datasets\R1_5d2w"
R1_out_dir = r"C:\Users\Rory Crean\Dropbox (lkgroup)\Backup_HardDrive\Postdoc\MachLearnConformationalFeatures\Workup\KE07_Workup\Raw_Results_Stats_Regress\R1_5d2w"

R1_pycontact_files_horizontal = ["PyContact_Per_Frame_Interactions_Block1.csv", "PyContact_Per_Frame_Interactions_Block2.csv",
                              "PyContact_Per_Frame_Interactions_Block3.csv", "PyContact_Per_Frame_Interactions_Block4.csv",
                              "PyContact_Per_Frame_Interactions_Block5.csv", "PyContact_Per_Frame_Interactions_Block6.csv",
                              "PyContact_Per_Frame_Interactions_Block7.csv", "PyContact_Per_Frame_Interactions_Block8.csv",
                              "PyContact_Per_Frame_Interactions_Block9.csv", "PyContact_Per_Frame_Interactions_Block10.csv",
                              "PyContact_Per_Frame_Interactions_Block11.csv", "PyContact_Per_Frame_Interactions_Block12.csv",
                              "PyContact_Per_Frame_Interactions_Block13.csv", "PyContact_Per_Frame_Interactions_Block14.csv",
                              "PyContact_Per_Frame_Interactions_Block15.csv", "PyContact_Per_Frame_Interactions_Block16.csv",
                              "PyContact_Per_Frame_Interactions_Block17.csv"]

R1_Trp50_Chi1_file = r"C:\Users\Rory Crean\Dropbox (lkgroup)\Backup_HardDrive\Postdoc\MachLearnConformationalFeatures\Workup\KE07_Workup\Raw_Datasets\R1_5d2w\R1_5d2w_1in10_Trp50_Chi1.dat"
R1_Trp50_Chi2_file = r"C:\Users\Rory Crean\Dropbox (lkgroup)\Backup_HardDrive\Postdoc\MachLearnConformationalFeatures\Workup\KE07_Workup\Raw_Datasets\R1_5d2w\R1_5d2w_1in10_Trp50_Chi2.dat"

# dataset prep parts. 
R1_pycontact_dataset = pycontact_processing.PyContactInitializer(
    pycontact_files=R1_pycontact_files_horizontal,
    multiple_files=True,
    merge_files_method="horizontal",  
    remove_false_interactions=True,
    in_dir=R1_in_dir,
)

# SPECIAL STEP FOR KEMP ELIMINASES BECAUSE PYCONTACT RENUMBERED THEM TO 0. 
R1_prepared_renumberd_df = pycontact_processing.modify_column_residue_numbers(
    dataset=R1_pycontact_dataset.prepared_df, 
    constant_to_add=1
)

R1_prepared_renumberd_df.head(3)
```

```python
# Now I need to add R1_Trp50_Chi1_file, and filter frames. 
df_chi1 = pd.read_csv(R1_Trp50_Chi1_file)
df_chi1 = df_chi1.set_axis(["Trp50Chi1"], axis=1)

R1_chi1_df = pd.concat([df_chi1, R1_prepared_renumberd_df], axis=1)
R1_chi1_df
```

```python
# Generate supervised dataset instance. 
R1_supervised_dataset = data_preperation.SupervisedFeatureData(
    input_df=R1_chi1_df,
    target_file=R1_Trp50_Chi2_file,
    is_classification=False,
    header_present=True 
)

R1_supervised_dataset.df_processed
```

```python
# Filtering 
R1_supervised_dataset.reset_filtering()
R1_supervised_dataset.filter_by_occupancy(min_occupancy=25)
R1_supervised_dataset.filter_by_avg_strength(average_strength_cut_off=1.0)
print(f"Number of features after filtering by average interaction scores: {len(R1_supervised_dataset.df_filtered.columns)}")
```

```python
# Now filter by Chi1 Range...
# remove all frames < 160 and > 240. 
R1_chi1_chi2_df = R1_supervised_dataset.df_filtered
R1_filtered_df = R1_chi1_chi2_df[( (R1_chi1_chi2_df.Trp50Chi1 > 160) & (R1_chi1_chi2_df.Trp50Chi1 < 240) )]
display("min = " + str(R1_filtered_df["Trp50Chi1"].min()))
display("max = " + str(R1_filtered_df["Trp50Chi1"].max()))
R1_filtered_df.head(3)
```

```python
# Now generate supervised dataset with Chi2
R1_df_ready = (R1_filtered_df.drop("Trp50Chi1", axis=1)).reset_index(drop=True) 
R1_df_ready.head(3)
```

```python
# Now time for stat regression analysis
R1_stat_model = stat_modelling.RegressionStatModel(
    dataset=R1_df_ready, 
    out_dir=R1_out_dir,
    interaction_types_included=["Hbond", "Saltbr", "Hydrophobic", "Other"] 
)
```

```python
R1_stat_model.scaled_dataset
```

```python
R1_stat_model.calc_mutual_info_to_target()
```

```python
R1_stat_model.calc_linear_correl_to_target()
```

```python
# First generate an instance of the class. 
R1_post_proc = post_proccessing.StatRegressorPostProcessor(
    stat_model=R1_stat_model,
    out_dir=R1_out_dir
)

# Now we can run the get_per_res_importance() method, changing the stat_method accordingly.
R1_mi_per_res_importances = R1_post_proc.get_per_res_importance(
    stat_method="mutual_information")
R1_lin_correl_per_res_importances = R1_post_proc.get_per_res_importance(
    stat_method="linear_correlation")
```

```python
# Write PyMOL compatable scripts for the per feature results.
# Simply swap between the two statistical methods as shown below. 
pymol_projections.project_pymol_top_features(
    per_feature_scores=R1_stat_model.linear_correlations,
    model_name="linear_correlation",
    numb_features=100,
    out_dir=R1_out_dir
)

pymol_projections.project_pymol_top_features(
    per_feature_scores=R1_stat_model.mutual_infos,
    model_name="mutual_information",
    numb_features=100,
    out_dir=R1_out_dir
)

# Write PyMOL compatable scripts for the per residue results.
# Simply swap between the two statistical methods as shown below. 
pymol_projections.project_pymol_per_res_scores(
    per_res_scores=R1_lin_correl_per_res_importances,
    model_name="linear_correlation",
    out_dir=R1_out_dir
)
pymol_projections.project_pymol_per_res_scores(
    per_res_scores=R1_mi_per_res_importances,
    model_name="mutual_information",
    out_dir=R1_out_dir
)
```
